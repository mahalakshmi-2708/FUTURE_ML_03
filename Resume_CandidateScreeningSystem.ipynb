```python
resumes = [
    "Python SQL Machine Learning Data Analysis",
    "Java Spring Boot Backend Development",
    "Python Excel Data Visualization",
    "C++ Embedded Systems"
]

job_description = "Python SQL Data Analysis Machine Learning"
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    return text

resumes_clean = [clean_text(r) for r in resumes]
job_clean = clean_text(job_description)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

all_text = resumes_clean + [job_clean]
vectors = vectorizer.fit_transform(all_text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

resume_vectors = vectors[:-1]
job_vector = vectors[-1]

scores = cosine_similarity(resume_vectors, job_vector)

print(scores)
```

```python
import numpy as np

ranked = sorted(
    list(enumerate(scores)),
    key=lambda x: x[1],
    reverse=True
)

for i, score in ranked:
    print(f"Resume {i+1} → Score: {score[0]:.2f}")
```

```python
job_skills = set(job_clean.split())

for i, resume in enumerate(resumes_clean):
    resume_skills = set(resume.split())
    missing = job_skills - resume_skills
    print(f"Resume {i+1} missing skills:", missing)
```
